1. Deep Learning dengan PyTorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
file_path = "Infrared.csv"  # Pastikan file ini ada dalam direktori kerja

# Membaca file CSV
df = pd.read_csv(file_path)

# Memilih fitur numerik untuk input dan target
features = ["T_atm", "Humidity", "Distance", "T_Max1", "T_OR1"]
target = "aveOralM"

# Menghapus baris dengan nilai NaN jika ada
df_clean = df.dropna(subset=features + [target])

# Pisahkan fitur (X) dan target (y)
X = df_clean[features].values
y = df_clean[target].values.reshape(-1, 1)

# Normalisasi data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Konversi ke tensor PyTorch
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Membagi data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

# Definisi model neural network
class InfraredNN(nn.Module):
    def __init__(self):
        super(InfraredNN, self).__init__()
        self.fc1 = nn.Linear(len(features), 8)  # Input -> Hidden 1
        self.fc2 = nn.Linear(8, 4)  # Hidden 1 -> Hidden 2
        self.fc3 = nn.Linear(4, 1)  # Hidden 2 -> Output

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)  # Output layer tanpa aktivasi (regresi)
        return x

# Inisialisasi model, loss function, dan optimizer
model = InfraredNN()
criterion = nn.MSELoss()  # Mean Squared Error untuk regresi
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training loop
epochs = 100
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

print("Training complete!")

# Evaluasi model
y_pred = model(X_test).detach().numpy()
y_true = y_test.numpy()

# Menghitung MSE, RMSE, dan R-Squared
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-Squared (R2): {r2}")
```

# penjelasan persamaan matematika PyTorch:

Loss function menunjukkan seberapa baik model belajar dari data, dengan nilai yang semakin kecil menandakan model makin akurat dalam memprediksi. Mean Squared Error (MSE) menghitung rata-rata kesalahan prediksi dengan cara mengkuadratkan selisih antara nilai asli dan hasil prediksi, di mana semakin kecil nilainya, semakin baik model bekerja. Root Mean Squared Error (RMSE) adalah versi MSE yang sudah diakar, sehingga lebih mudah dipahami karena memiliki satuan yang sama dengan data aslinya, dan semakin kecil RMSE, semakin baik kualitas prediksi model. R-Squared (R²) mengukur sejauh mana model dapat menjelaskan variasi dalam data, dengan nilai 1 berarti model sangat baik, sementara nilai negatif yang sangat besar menunjukkan model justru buruk dalam memprediksi. Dari hasil ini, meskipun loss terus menurun, nilai R² yang sangat rendah menandakan model gagal memahami pola dalam data, kemungkinan akibat overfitting, pemilihan fitur yang kurang tepat, atau masalah dalam pengolahan data.

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
csv_path = "/content/Infrared.csv"
df = pd.read_csv(csv_path)

# Handle missing values (drop or fill with mean)
df = df.dropna()  # Menghapus baris dengan NaN (opsi lain: df.fillna(df.mean(), inplace=True))

# Select features (excluding non-numeric columns)
X = df.select_dtypes(include=['float64', 'int64']).drop(columns=['aveOralM'])
y = df['aveOralM']

# Convert to TensorFlow tensors
X_tf = tf.convert_to_tensor(X.values, dtype=tf.float32)
y_tf = tf.convert_to_tensor(y.values.reshape(-1, 1), dtype=tf.float32)

# Define model
model_tf = Sequential([
    Dense(16, activation='relu', input_shape=(X.shape[1],)),
    Dense(8, activation='relu'),
    Dense(1, activation='linear')  # Regression output
])

# Compile model
model_tf.compile(loss='mse', optimizer=SGD(learning_rate=0.01), metrics=['mae'])

# Train model
model_tf.fit(X_tf, y_tf, epochs=100, verbose=1)

# Predict values
y_pred = model_tf.predict(X_tf)

# Compute evaluation metrics
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R-Squared: {r2}")
```

Pada setiap epoch, model belajar untuk mengurangi kesalahan dalam memprediksi data.
